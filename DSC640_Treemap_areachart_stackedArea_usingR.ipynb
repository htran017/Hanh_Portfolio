```r
unemploy_df <- read.csv('unemploymentrate1948_2010.csv')
```

```r
head(unemploy_df)
```

```r
library(ggplot2)
```

```r
plot <- ggplot(data=unemploy_df, aes(x=unemploy_df$Year, y=unemploy_df$Value, fill=unemploy_df$Period))
```

```r
plotgeom <- plot + geom_area(aes(colour=unemploy_df$Period, fill= unemploy_df$Period))
```

```r
plotgeom + ggtitle("Unemployment from 1948-2010") +labs(x="Year", y="Unemployment")
```

```r
# Area chart
ggplot(unemploy_df, aes(x=Year, y=Value)) +
  geom_area( fill="#69b3a2", alpha=0.4) +
  geom_line(color="#69b3a2", size=2) +
  geom_point(size=3, color="#69b3a2") +
  ggtitle("Unemployment 1948-2010")
```
