```python
import pandas as pd
import numpy as np
```

## Finding Movie Similarities based on Correlation

```python
r_cols = ['userId', 'movieId', 'rating']
ratings = pd.read_csv('ratings.csv', names=r_cols, usecols=range(3))
```

```python
m_cols = ['movieId', 'title']
movies = pd.read_csv('movies.csv', names=m_cols, usecols=range(2))
```

```python
#merge the datasets containing movie title and movie id, ratings, & userid
ratings = pd.merge(movies, ratings)
```

```python
ratings.head()
```

```python
#drop the extra row containing duplicate column names
ratings = ratings.drop(labels=0, axis=0)
```

```python
ratings.head()
```

```python
#check datatype for datatype error
dataTypeSeries = ratings.dtypes
print(dataTypeSeries)
```

```python
#convert datatype to numeric
ratings['movieId'] = pd.to_numeric(ratings['movieId'])
```

```python
ratings['userId'] = pd.to_numeric(ratings['userId'])
```

```python
ratings['rating'] = pd.to_numeric(ratings['rating'])
```

```python
#use pivot table to construct a user/movie rating matrix. 
movieRatings = ratings.pivot_table(index=['userId'], columns=['title'], values = 'rating')
```

```python
#The NaN indicates missing data - movies that specific users didn't rate.
movieRatings.head()
```

```python
#check a specific movie similarity such as for Fight Club
fightClubRatings = movieRatings['Fight Club (1999)']
fightClubRatings.head()
```

```python
#based on rating find a correlation between movies
similarMovies = movieRatings.corrwith(fightClubRatings)
similarMovies = similarMovies.dropna()
df = pd.DataFrame(similarMovies)
df.head(10)
```

```python
#sort for most similar movies to Fight Club
similarMovies.sort_values(ascending=False)
```

```python
#check the size or number of ratings for each movie
movieStats = ratings.groupby('title').agg({'rating': [np.size, np.mean]})
movieStats.head()
```

```python
#get rid of spurious data where there are less than 200 ratings for a given movie 
popularMovies = movieStats['rating']['size'] >=200
movieStats[popularMovies].sort_values([('rating', 'mean')], ascending=False)[:20]
```

```python
#add similarity correlation column
df = movieStats[popularMovies].join(pd.DataFrame(similarMovies, columns=['similarity']))
df.head(15)
```

```python
#sort similarity to get the highest correlation on top to show the most similar movies to Fight Club
df.sort_values(['similarity'], ascending=False) [:15]
```

## Item- Based Collaborative Filtering

```python
corrMatrix = movieRatings.corr(method='pearson', min_periods=100)
```

```python
corrMatrix.head()
```

```python
#Create a sim User to test recommendation system. User must provide a few initial ratings 
myRatings = movieRatings.loc[0].dropna()
myRatings
```

```python
simCandidates = pd.Series()
for i in range(0, len(myRatings.index)):
    print("Adding sims for " + myRatings.index[i] + "...")
    sims = corrMatrix[myRatings.index[i]].dropna()
    sims = sims.map(lambda x: x * myRatings[i])
    simCandidates = simCandidates.append(sims)
    
print("sorting...")
simCandidates.sort_values(inplace=True, ascending=False)
print(simCandidates.head(10))
```

```python
simCandidates =simCandidates.groupby(simCandidates.index).sum()
```

```python
simCandidates.sort_values(inplace=True, ascending=False)
simCandidates.head(25)
```
